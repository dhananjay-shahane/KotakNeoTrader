```python
# ============================================================================
# CELL 10: Place Order Functions
# ============================================================================

def place_market_order(trading_symbol, quantity, transaction_type, product="CNC", 
                      exchange_segment="nse_cm"):
    
    """Place a market order"""
    try:
        print(f"üìã Placing {transaction_type} market order for {quantity} shares of {trading_symbol}")
        
        order_response = client.place_order(
            exchange_segment=exchange_segment,
            product=product,
            price="0",  # Market order - price is 0
            order_type="MKT",
            quantity=str(quantity),
            validity="DAY",
            trading_symbol=trading_symbol,
            transaction_type=transaction_type,
            amo="NO",
            disclosed_quantity="0",
            market_protection="0",
            pf="N",
            trigger_price="0",
            tag=f"API_ORDER_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        )
        
        print("‚úÖ Market order placed successfully!")
        print(f"Order Response: {order_response}")
        
        return order_response
        
    except Exception as e:
        print(f"‚ùå Error placing market order: {str(e)}")
        return None

def place_limit_order(trading_symbol, quantity, price, transaction_type, 
                     product="CNC", exchange_segment="NSE"):
    """Place a limit order"""
    try:
        print(f"üìã Placing {transaction_type} limit order for {quantity} shares of {trading_symbol} at ‚Çπ{price}")
        
        order_response = client.place_order(
            exchange_segment=exchange_segment,
            product=product,
            price=str(price),
            order_type="L",
            quantity=str(quantity),
            validity="DAY",
            trading_symbol=trading_symbol,
            transaction_type=transaction_type,
            amo="NO",
            disclosed_quantity="0",
            market_protection="0",
            pf="N",
            trigger_price="0",
            tag=f"API_ORDER_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        )
        
        print("‚úÖ Limit order placed successfully!")
        print(f"Order Response: {order_response}")
        
        return order_response
        
    except Exception as e:
        print(f"‚ùå Error placing limit order: {str(e)}")
        return None

def place_stop_loss_order(trading_symbol, quantity, price, trigger_price, 
                         transaction_type, product="CNC", exchange_segment="nse_cm"):
    """Place a stop loss order"""
    try:
        print(f"üìã Placing {transaction_type} stop loss order for {quantity} shares of {trading_symbol}")
        print(f"Price: ‚Çπ{price}, Trigger Price: ‚Çπ{trigger_price}")
        
        order_response = client.place_order(
            exchange_segment=exchange_segment,
            product=product,
            price=str(price),
            order_type="SL",
            quantity=str(quantity),
            validity="DAY",
            trading_symbol=trading_symbol,
            transaction_type=transaction_type,
            amo="NO",
            disclosed_quantity="0",
            market_protection="0",
            pf="N",
            trigger_price=str(trigger_price),
            tag=f"API_ORDER_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        )
        
        print("‚úÖ Stop loss order placed successfully!")
        print(f"Order Response: {order_response}")
        
        return order_response
        
    except Exception as e:
        print(f"‚ùå Error placing stop loss order: {str(e)}")
        return None
```

```python
# ============================================================================
# CELL 11: Interactive Order Placement
# ============================================================================

def interactive_order_placement():
    """Interactive order placement interface"""
    if not login_success:
        print("‚ùå Please login first to place orders.")
        return
    
    print("\n" + "="*60)
    print("üìã INTERACTIVE ORDER PLACEMENT")
    print("="*60)
    
    try:
        # Get order details from user
        trading_symbol = input("Enter trading symbol (e.g., RELIANCE): ").strip().upper()
        quantity = int(input("Enter quantity: "))
        
        print("\nTransaction Type:")
        print("B - Buy")
        print("S - Sell")
        transaction_type = input("Enter transaction type (B/S): ").strip().upper()
        
        print("\nOrder Type:")
        print("1. Market Order")
        print("2. Limit Order")
        print("3. Stop Loss Order")
        order_type_choice = input("Enter choice (1/2/3): ").strip()
        
        print("\nProduct Type:")
        print("CNC - Cash and Carry")
        print("MIS - Margin Intraday")
        print("NRML - Normal")
        product = input("Enter product type (CNC/MIS/NRML): ").strip().upper()
        
        print("\nExchange Segment:")
        print("nse_cm - NSE Cash Market")
        print("bse_cm - BSE Cash Market")
        exchange_segment = input("Enter exchange segment (default: nse_cm): ").strip().lower()
        if not exchange_segment:
            exchange_segment = "nse_cm"
        
        # Place order based on type
        if order_type_choice == "1":
            # Market Order
            order_response = place_market_order(
                trading_symbol, quantity, transaction_type, product, exchange_segment
            )
        elif order_type_choice == "2":
            # Limit Order
            price = float(input("Enter limit price: "))
            order_response = place_limit_order(
                trading_symbol, quantity, price, transaction_type, product, exchange_segment
            )
        elif order_type_choice == "3":
            # Stop Loss Order
            price = float(input("Enter price: "))
            trigger_price = float(input("Enter trigger price: "))
            order_response = place_stop_loss_order(
                trading_symbol, quantity, price, trigger_price, transaction_type, product, exchange_segment
            )
        else:
            print("‚ùå Invalid choice!")
            return
        
        if order_response:
            print("\n‚úÖ Order placed successfully!")
            return order_response
        else:
            print("\n‚ùå Order placement failed!")
            return None
            
    except ValueError as e:
        print(f"‚ùå Input error: {str(e)}")
        return None
    except Exception as e:
        print(f"‚ùå Error in order placement: {str(e)}")
        return None

# Example usage - Uncomment to use interactively
# order_response = interactive_order_placement()
```

```python
# ============================================================================
# CELL 14: Utility Functions
# ============================================================================

def get_market_quotes(trading_symbol, exchange_segment="nse_cm"):
    """Get market quotes for a symbol"""
    try:
        # First search for the instrument token
        search_result = client.search_scrip(
            exchange_segment=exchange_segment,
            symbol=trading_symbol,
            expiry="",
            option_type="",
            strike_price=""
        )
        
        if not search_result or not search_result.get('data'):
            print(f"‚ùå Symbol '{trading_symbol}' not found")
            return None
        
        # Get the first matching result
        instrument_data = search_result['data'][0]
        instrument_token = instrument_data['instrument_token']
        
        print(f"üìä Getting quotes for {trading_symbol}...")
        
        instrument_tokens = [{
            "instrument_token": instrument_token,
            "exchange_segment": exchange_segment
        }]
        
        quotes = client.quotes(instrument_tokens=instrument_tokens, quote_type="all")
        
        if quotes:
            print("‚úÖ Quotes retrieved successfully!")
            print(f"Quotes: {json.dumps(quotes, indent=2)}")
            return quotes
        else:
            print("‚ùå Failed to retrieve quotes.")
            return None
            
    except Exception as e:
        print(f"‚ùå Error fetching quotes: {str(e)}")
        return None

# def logout():
#     """Logout from the session"""
#     try:
#         print("üëã Logging out...")
#         logout_response = client.logout()
#         print("‚úÖ Logged out successfully!")
#         print(f"Logout Response: {logout_response}")
#         return logout_response
#     except Exception as e:
#         print(f"‚ùå Error during logout: {str(e)}")
#         return None
```
